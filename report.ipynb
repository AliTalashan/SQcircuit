````python
loop1 = sq.Loop()

C = sq.Capacitor(15, 'GHz', Q=1.5e6)
CJ = sq.Capacitor(0.1, 'GHz', Q=1e6)
Csh = sq.Capacitor(0.2, 'GHz')
L = sq.Inductor(0.15, 'GHz')
JJ = sq.Junction(8, 'GHz', cap=CJ)
````

# **۲. گرفتن خروجی و دیسکریپشن مدار**




پیش از این با نحوه نصب کتابخانه SQ و چگونگی تعریف مدار با استفاده از متدهای این
کتابخانه آشنا شدیم. قدم بعدی پس از تعریف مدار، بررسی ویژگی‌های مدار است. برای این منظور کتابخانه  SQ متدی را تحت عنوان description معرفی می‌کند که برای شناخت اولیه مدارهای تعریف شده، بسیار کار آمد است.
برای اینکه شناخت بهتری از این متد و خروجی آن داشته باشیم، آن را روی مدار JC اعمال می‌کنیم:
</div>


````python
loop_jc = sq.Loop()
JJ_jc = sq.Junction(100, 'GHz', loops=[loop_jc])

elements = {(0, 1): [JJ_jc],
            (1, 0): [C]
            }

jc = sq.Circuit(elements)
jc.description()
````

<div dir=rtl>

همانطور که مشاهده می‌کنید، متد description هامیلتونی مدار را به همراه لیستی از مودها (در این مدار فقط charge mode داریم اما در حالت کلی هم charge mode و هم مودهای هارمونیک)،‌ بر می‌گرداند. همچنین مقادیر شار خروجی و پارامترهای همیلتونی را نیز گزارش می‌کند.

مقدار ng_1 در مود 1 را نیز که بیانگر normalized charge offset است می‌توان به صورت دستی با استفاده از متد زیر تغییر داد:

<div dir=ltr>

```
cr.set_charge_offset(mode=1, ng=#number)
```

این متد برای فهم پراکندگی بار اسپکتروم انرژی که در اثر انحرافات رندوم در charge island ها به وجود می‌آید، بسیار کاربردی است.

می‌توانیم با استفاده از این متد آن تعدادی از ویژه حالت‌های مدار که نیاز داریم را پیدا کنیم:

<div dir=rtl>

برای رسم اسپکتروم مدار، باید شار خارجی loop را با استفاده از متد `set_flux` تغییر دهیم و سپس ویژه فرکانس‌های متناظر با این شار خارجی را بیابیم. با استفاده از کد زیر به ازای شارهای خارجی مختلف، ویژه فرکانس‌هارا استخراج و نمودار متناظر با آن‌ها را رسم می‌کنیم.

````python
# external flux for sweeping over
phi = np.linspace(0,1,100)

# spectrum of the JC circuit
spec_jc = np.zeros((n_eig, len(phi)))

for i in range(len(phi)):
    # set the external flux for the loop
    loop_jc.set_flux(phi[i])

    # diagonlize the circuit
    spec_jc[:, i], _ = jc.diag(n_eig)

plt.figure()
for i in range(n_eig):
    if i == 0:
      plt.plot(phi, spec_jc[i,:]- spec_jc[0,:], label='JC')
      plt.legend(handletextpad=-0.1, handlelength=0.0)
    else:
      plt.plot(phi, spec_jc[i,:]- spec_jc[0,:])

plt.xlabel(r"$\Phi_{ext}/\Phi_0$", fontsize=13)
plt.ylabel(r"$f_i-f_0$[GHz]", fontsize=13)
plt.show()
````

````python
# external flux for sweeping over
phi = np.linspace(0,1,100)

# spectrum of the LC circuit
n_eig=5
spec_lc = np.zeros((n_eig, len(phi)))

for i in range(len(phi)):
    # set the external flux for the loop
    loop_lc.set_flux(phi[i])

    # diagonlize the circuit
    spec_lc[:, i], _ = lc.diag(n_eig)


plt.figure()
for i in range(n_eig):
    if i == 0:
      plt.plot(phi, spec_lc[i,:]- spec_lc[0,:], label='LC')
      plt.legend(handletextpad=-0.1, handlelength=0.0)
    else:
      plt.plot(phi, spec_lc[i,:]- spec_lc[0,:])

plt.xlabel(r"$\Phi_{ext}/\Phi_0$", fontsize=13)
plt.ylabel(r"$f_i-f_0$[GHz]", fontsize=13)
plt.show()
````

# **۴. اثرات اختلالی و نویزها**
 در بخش‌های قبل به تعریف مدار و گرفتن خروجی‌های مختلف برای آن توسط کتابخانه  SQcircuit  پرداختیم. حال در این بخش قطعات غیر ایده‌آل و ورودی‌های نویزدار را به کتابخانه معرفی میکنیم و نحوه مشاهده اثرات غیرایده‌آلی و گرفتن خروجی‌های مربوط به آن را بازگو می‌کنیم.
 در این بخش از پرداختن به جزئیات ریاضی مسائل خودداری میکنیم و بیشتر به نحوه اجرای آن در کتابخانه میپردازیم. برای مطالعه بیشتر در این مورد میتوانید به بخش ۵.۳ مقاله اصلی کتابخانه مراجعه کنید.

<div dir=rtl>


## تعریف اختلال‌ها:

 با نوشتن محاسبات و وارد کردن اثرات اختلالی به قطعات، در نهایت برای هر کدام از قطعات خازن و القاگر میتوان کمیتی با نام معیار کیفیت (Quality Factor) معرفی کرد که سنجه‌ای برای اندازه‌گیری ایده‌آل بودن آن‌هاست و میتواند تابعی از فرکانس نسبی بین ترازهای انرژی باشد. این کمیت را با Q نشان میدهیم و مقدار پیش‌فرض آن‌ها در کتابخانه SQcircuit به شکل زیر است:


<div dir=ltr>


> $Q_{cap}(w) = 10^6 (\frac{2π  × 6GHz}{|w|})^{0.7}$


> $Q_{ind}(w)= 500 × 10^6 \frac{K_0(\frac{h × 0.5GHz}{2k_bT}). sinh(\frac{h × 0.5GHz}{2k_bT})}{K_0(\frac{h|w|}{2k_bT}). sinh(\frac{h|w|}{2k_bT})}$



<div dir=rtl>


  اگر بخواهیم مقدار پیش‌فرض آن‌ها را به دلخواه عوض کنیم، میبایست هنگام تعریف‌ المان اندازه آن را مشخص کنیم:

<div dir=ltr>

```
C = sq.Capacitor(15, 'GHz', Q=1.5e6)
L = sq.Inductor(0.15, 'GHz', loops=[loop1])
```

 برای جوزفسون جانکشن نیز دو کمیت برای تعریف اختلال داریم: $∆$ شکاف ابررسانایی (superconductive gap)  و $x_{qp}$ چگالی شبه‌ذرات (quasiparticle density). مقدار پیش‌فرض این دو کمیت به شکل زیر است:



<div dir=ltr>


> $Δ=3.4 × 10^{-4} \space eV$

> $x_{qp}=3 × 10^{-6}$


<div dir=rtl>


 و برای مشخص کردن این کمیت‌ها به طور دستی نیز مانند خازن و القاگر به شکل زیر عمل میکنیم:

##آهنگ هدررفت $(decay \space rate)$

بالاخره و با تعریف تمام عناصر هدررفت، میتوانیم decay rate یا آهنگ هدررفت در گذار انرژی بین حالت $m$ و $n$ را محاسبه کنیم. این کمیت در بخش ۵.۳ مقاله به شکل زیر آمده


<div dir=ltr>


>$Γ_{m→n}=\frac{1}{ħ^2}|⟨m|\hat{\mathcal{O}}|n⟩|^2S_{ξξ}(w_{mn})\quad\quad\quad\quad\quad (eq.27)$

<div dir=rtl>

 و طبق آن زمان خارج شدن از حالت پولاریزه در کیوبیت به دست می‌آید
 برای محاسبه این تابع در کتابخانه SQcircuit از متد dec_rate کمک میگیریم

<div dir=ltr>

```
Gamma_mn = cr.dec_rate(dec_type=dec_type,
                       states=(m, n),
                       total=total)
```

<div dir=rtl>

 در کد بالا اگر total=False باشد،‌ کتابخانه فقط هدر رفت در گذار لایه بالاتر به پایین‌تر را محاسبه میکند، در حالی که اگر total=True باشد (که حالت پیش‌فرض کتابخانه نیز هست) جمع $\space Γ_{m→n} + Γ_{n→m}$ در نظر گرفته میشود

 همچنین جمله dec_type نوع هدر رفت را مشخص میکند. به ازای قطعات غیر ایده‌آل مختلف، این آرگومان میتواند یکی از ۳ حالت "capacitive", "inductive" و یا "quasiparticle" را داشته باشد.

 همچنین برای محاسبه خطای ایجاد شده به واسطه نویزها نیز ساز و کار مشابهی وجود دارد. در این حالت تابعی به نام $κ_{mn}$ تعریف میکنیم که رابطه زیر برای آن برقرار است

<div dir=ltr>

> $κ_{mn} = \frac{1}{T_\phi}$

<div dir=rtl>

 و همچنین به طریق مشابهی و به کمک متد dec_rate به دست می‌آید که در آن dec_type مقادیر "flux", "charge" و یا "cc" (برای حالت جریان بحرانی یا critical current) را میگیرد


 بنابراین حالا با داشتن آهنگ هدررفت، میتوانیم طول عمر کیوبیت و اطلاعات در آن را به کمک این کتابخانه نمایش دهیم.


<div dir=rtl>


## Depolarization & Dephasing times

 حال با استفاده از روابطی که برای $T_1$ و $T_ϕ$ داریم، آنها را میتوانیم با استفاده از خروجی متد dec_rate برای شارهای ورودی مختلف به دست آوریم. در ادامه با ذکر یک مثال این عمل را انجام میدهیم

</div>

````python
loop1 = sq.Loop()

L = sq.Inductor(0.15, 'GHz', loops=[loop1])
JJ = sq.Junction(8, 'GHz', loops=[loop1])

elements = {(0,1): [C, L, JJ]}
lcj_dec = sq.Circuit(elements)
lcj_dec.description()
lcj_dec.set_trunc_nums([20])
n_eig = 2
````
